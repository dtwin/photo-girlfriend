```python
from PIL import Image
import os.path
import glob
def convertjpg(jpgfile,outdir,width=128,height=128):
    img=Image.open(jpgfile)
    try:
        new_img=img.resize((width,height),Image.BILINEAR)   
        new_img.save(os.path.join(outdir))
    except Exception as e:
        print(e)
count = 2030        
for i in range(1000):

    link = "C:\\resize\\{}.jpg".format(count)
    convertjpg("C:\\photo1\\367.jpg",link)
    count = count + 1
    print('ok')
print(count)
```
